# Exploratory Data Analysis - creditcard.csv

This notebook performs exploratory data analysis on the creditcard.csv dataset.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

# Load the dataset
creditcard_path = '../data/raw/creditcard.csv'
credit_df = pd.read_csv(creditcard_path)

# Display first few rows
credit_df.head()
```

```python
# Basic information
print("Shape:", credit_df.shape)
print("\nData types:")
print(credit_df.dtypes)
print("\nMissing values:")
print(credit_df.isnull().sum())
print("\nSummary statistics:")
credit_df.describe()

# Class distribution
plt.figure(figsize=(8, 6))
credit_df['Class'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Count')
plt.show()

print("Class distribution:")
print(credit_df['Class'].value_counts())
print("\nPercentage:")
print(credit_df['Class'].value_counts(normalize=True) * 100)
```

```python
# Analyze Amount and Time
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Amount distribution
sns.histplot(credit_df['Amount'], kde=True, ax=axes[0,0])
axes[0,0].set_title('Distribution of Amount')

sns.boxplot(x=credit_df['Amount'], ax=axes[0,1])
axes[0,1].set_title('Boxplot of Amount')

# Time distribution
sns.histplot(credit_df['Time'], kde=True, ax=axes[1,0])
axes[1,0].set_title('Distribution of Time')

sns.boxplot(x=credit_df['Time'], ax=axes[1,1])
axes[1,1].set_title('Boxplot of Time')

plt.tight_layout()
plt.show()

# Amount vs Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title('Amount vs Class')
plt.show()
```

```python
# Save processed data (minimal processing for creditcard)
processed_path = '../data/processed/creditcard_processed.csv'
credit_df.to_csv(processed_path, index=False)
print(f"Processed data saved to {processed_path}")
```
